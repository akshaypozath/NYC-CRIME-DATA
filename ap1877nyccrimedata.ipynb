```python

def get_borough(precinct):
    if 1 <= precinct <= 17:
        return "Manhattan South"
    elif 19 <= precinct <= 34:
        return "Manhattan North"
    elif 40 <= precinct <= 52:
        return "Bronx"
    elif 60 <= precinct <= 78:
        return "Brooklyn South"
    elif 73 <= precinct <= 94:
        return "Brooklyn North"
    elif precinct in [100, 101, 102, 103, 105, 106, 107, 113]:
        return "Queens South"
    elif precinct in [104, 108, 109, 110, 111, 112, 114, 115]:
        return "Queens North"
    elif 120 <= precinct <= 123:
        return "Staten Island"
    else:
        return "Unknown" 

data['Borough'] = data['PCT'].apply(get_borough)

updated_file_path = "updated_cleanedcrimedata.csv"
data.to_csv(updated_file_path, index=False)

print(data.head())

#this function separates each of the precincts by borough based on the above conditions
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "updated_cleanedcrimedata.csv"  
data = pd.read_csv(file_path)


year_columns = [str(year) for year in range(2000, 2024)]

long_data = data.melt(
    id_vars=['PCT', 'CRIME', 'Borough'],
    value_vars=year_columns,
    var_name='Year',
    value_name='Count'
)

long_data['Year'] = long_data['Year'].astype(int)

data_2023 = long_data[long_data['Year'] == 2023]

crime_borough_2023 = data_2023.groupby(['Borough', 'CRIME'])['Count'].sum().reset_index()

highest_crime_boroughs = []

for crime in crime_borough_2023['CRIME'].unique():
    crime_data = crime_borough_2023[crime_borough_2023['CRIME'] == crime]
    max_borough = crime_data.loc[crime_data['Count'].idxmax()] 
    highest_crime_boroughs.append({
        'Crime': crime,
        'Borough': max_borough['Borough'],
        'Count': max_borough['Count']
    })

print("Borough with the Highest Crime Count for Each Crime Type in 2023:")
for result in highest_crime_boroughs:
    print(f"Crime: {result['Crime']}, Borough: {result['Borough']}, Count: {result['Count']}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "updated_cleanedcrimedata.csv" 
data = pd.read_csv(file_path)

manhattan_south_data = data[
    (data['Borough'] == "Manhattan South") & (data['CRIME'] != "TOTAL SEVEN MAJOR FELONY OFFENSES")
]

year_columns = [str(year) for year in range(2000, 2024)]

long_data = manhattan_south_data.melt(
    id_vars=['PCT', 'CRIME', 'Borough'],
    value_vars=year_columns,
    var_name='Year',
    value_name='Count'
)

long_data['Year'] = long_data['Year'].astype(int)

mean_data = long_data.groupby(['CRIME', 'Year'])['Count'].mean().reset_index()

plt.figure(figsize=(12, 8))

for crime in mean_data['CRIME'].unique():
    crime_data = mean_data[mean_data['CRIME'] == crime]
    plt.plot(crime_data['Year'], crime_data['Count'], label=crime)

plt.title("Mean Crime Trends in Manhattan South (2000-2023)", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Mean Crime Count (Across All Precincts)", fontsize=12)
plt.legend(title="Crime Type", fontsize='small', loc='upper left')
plt.grid(visible=True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

The above code takes the mean of all the precincts in the Manhattan South Borough for each year, for each of the 6 crime types. This allows us to analyze the historical trends from 2000 to 2023. It seems as if murder and rape have remained relatively stable over the 23 years analyzed. However, at around the early 2020's we can see an incline in the other crime types although they were preceded by a consistent decline in the years preciding it.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


file_path = "updated_cleanedcrimedata.csv"  
data = pd.read_csv(file_path)

data = data[(data['CRIME'] != "TOTAL SEVEN MAJOR FELONY OFFENSES") & (data['Borough'] != "Unknown")]

year_columns = [str(year) for year in range(2000, 2024)]

long_data = data.melt(
    id_vars=['PCT', 'CRIME', 'Borough'],
    value_vars=year_columns,
    var_name='Year',
    value_name='Count'
)

long_data['Year'] = long_data['Year'].astype(int)

mean_data = long_data.groupby(['Borough', 'CRIME', 'Year'])['Count'].mean().reset_index()

for borough in mean_data['Borough'].unique():
    borough_data = mean_data[mean_data['Borough'] == borough]
    plt.figure(figsize=(12, 8)) 
    prediction_results = []

    
    for crime in borough_data['CRIME'].unique():
        crime_data = borough_data[borough_data['CRIME'] == crime]
        
        crime_data['PercentChange'] = crime_data['Count'].pct_change()

        avg_percent_change = crime_data['PercentChange'].mean()

        last_year = crime_data['Year'].max()
        last_count = crime_data['Count'].iloc[-1]
        future_years = range(last_year + 1, last_year + 6)
        future_predictions = []
        
        current_value = last_count
        for year in future_years:
            current_value *= (1 + avg_percent_change)  
            future_predictions.append(current_value)
            prediction_results.append({
                "Borough": borough,
                "Crime": crime,
                "Year": year,
                "Predicted Count": round(current_value, 2)
            })
        
        plt.plot(crime_data['Year'], crime_data['Count'], label=f"{crime} (Historical)", marker='o')
        
        plt.plot(
            future_years, future_predictions,
            label=f"{crime} (Prediction)", linestyle='--'
        )
    
    plt.title(f"Mean Crime Trends with Percent Change Predictions ({borough})", fontsize=16)
    plt.xlabel("Year", fontsize=12)
    plt.ylabel("Mean Crime Count (Across All Precincts)", fontsize=12)
    plt.legend(title="Crime Type", fontsize='small', loc='upper left')
    plt.grid(visible=True, linestyle='--', alpha=0.5)
    
    plt.tight_layout()
    plt.show()

print("Predicted Crime Numbers for All Boroughs (2024-2028):")
for result in prediction_results:
    print(f"Borough: {result['Borough']}, Crime: {result['Crime']}, Year: {result['Year']}, Predicted Count: {result['Predicted Count']}")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np


train_data = mean_data[mean_data['Year'] <= 2018]
test_data = mean_data[mean_data['Year'] > 2018]

rmse_results = []
mae_results = []

for borough in mean_data['Borough'].unique():
    borough_train = train_data[train_data['Borough'] == borough]
    borough_test = test_data[test_data['Borough'] == borough]
    
    for crime in borough_train['CRIME'].unique():
        crime_train = borough_train[borough_train['CRIME'] == crime]
        crime_test = borough_test[borough_test['CRIME'] == crime]
        
        crime_train.loc[:, 'PercentChange'] = crime_train['Count'].pct_change()
        

        crime_train = crime_train.replace([np.inf, -np.inf], np.nan)  
        crime_train = crime_train.dropna(subset=['PercentChange'])  
        
        
        avg_percent_change = crime_train['PercentChange'].mean()
        
       
        current_value = crime_train['Count'].iloc[-1]
        predictions = [] 
        for year in crime_test['Year']:
            current_value *= (1 + avg_percent_change)  
            predictions.append(current_value)
        
       
        actuals = crime_test['Count'].values
        
        valid_indices = ~np.isnan(predictions) & ~np.isnan(actuals)  
        predictions_clean = np.array(predictions)[valid_indices]  
        actuals_clean = actuals[valid_indices]  
        
        if len(predictions_clean) == len(actuals_clean):
            rmse = mean_squared_error(actuals_clean, predictions_clean, squared=False)
            mae = mean_absolute_error(actuals_clean, predictions_clean)
            rmse_results.append({
                'Borough': borough,
                'Crime': crime,
                'RMSE': rmse,
                'MAE': mae
            })
        else:
            print(f"Warning: Length incorrect for crime '{crime}' in borough '{borough}'.")

print("RMSE and MAE for each Crime Type in each Borough:")
for result in rmse_results:
    print(f"Borough: {result['Borough']}, Crime: {result['Crime']}, RMSE: {result['RMSE']:.2f}, MAE: {result['MAE']:.2f}")
```

As we can see by the results above, the model does decent with very common crime types such as grand larceny but does extremely well with forecasting crime types such as rape and murder due to the fact that relative to the others they occur at a less frequency.
